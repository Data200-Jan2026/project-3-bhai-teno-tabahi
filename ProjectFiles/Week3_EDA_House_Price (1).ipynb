```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv"
df = pd.read_csv(url)

df.head()
```

The housing dataset was loaded into Google Colab using the Pandas library for exploratory data analysis.

```python
df.shape
```

The dataset contains multiple rows and columns representing housing features and prices.

```python
df.info()

```

```python
df.describe()
```

Descriptive statistics were used to summarize numerical variables such as median house value, median income, and total rooms. This helped in understanding the distribution and range of the data.

```python
df.isnull().sum()
```

```python
median_bedrooms = df['total_bedrooms'].median()
df['total_bedrooms'] = df['total_bedrooms'].fillna(median_bedrooms)

```

```python
df.isnull().sum()
```

Missing values were identified in the total_bedrooms variable. These missing values were handled by replacing them with the median value to avoid data loss and reduce the effect of outliers.

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

Duplicate records were checked and removed to ensure data consistency and avoid biased statistical analysis.

```python
plt.figure()
plt.boxplot(df['median_house_value'])
plt.ylabel("House Price")
plt.title("Box Plot of Median House Value")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['total_rooms'])
plt.ylabel("Total Rooms")
plt.title("Box Plot of Total Rooms")
plt.show()
```

Box plots were used to identify outliers in house prices and total rooms. Extreme values were observed, indicating the presence of outliers in the dataset.

```python
plt.figure()
plt.hist(df['median_house_value'], bins=30)
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.title("Distribution of House Prices")
plt.show()
```

```python
plt.figure()
plt.hist(df['median_income'], bins=30)
plt.xlabel("Median Income")
plt.ylabel("Frequency")
plt.title("Distribution of Median Income")
plt.show()
```

```python
plt.figure()
plt.scatter(df['median_income'], df['median_house_value'])
plt.xlabel("Median Income")
plt.ylabel("House Price")
plt.title("Median Income vs House Price")
plt.show()
```

**Key Insights**

The distribution of house prices is right-skewed, indicating that most houses fall within a moderate price range while a few are highly priced. Median income shows a strong positive relationship with house prices, meaning areas with higher income tend to have higher house values. Outliers are present in the dataset, as observed through box plots. Overall, income and housing characteristics significantly influence house prices.
